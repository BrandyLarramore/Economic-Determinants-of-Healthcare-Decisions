```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path_v = '../data/Vaccination_Coverage_among_Adults__18__Years__20240107.csv'
vaccination_data = pd.read_csv(file_path_v)
vaccination_data.head()
```

```python
vaccination_data.describe()
```

```python
vaccination_data["Survey Year"].max()
```

```python
vaccination_data['Estimate (%)'] = pd.to_numeric(vaccination_data['Estimate (%)'], errors='coerce')
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Vaccine', y='Estimate (%)', data=vaccination_data)
plt.title('Vaccination Rates by Type')
plt.xlabel('Vaccine Type')
plt.ylabel('Vaccination Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='Survey Year', y='Estimate (%)', hue='Vaccine', data=vaccination_data, errorbar=None)
plt.title('Vaccination Rates Over Time')
plt.xlabel('Survey Year')
plt.ylabel('Vaccination Rate (%)')
plt.legend(title='Vaccine', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Estimate (%)', y='Geography', data=vaccination_data, ci=None)
plt.title('Average Vaccination Rates by Location')
plt.xlabel('Average Vaccination Rate (%)')
plt.ylabel('Geography')
plt.show()
```
