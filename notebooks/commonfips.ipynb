```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file_path_i = '../data/income2019.csv'
income_data = pd.read_csv(file_path_i, skiprows=[0])
income_data.head
```

```python
income_data.info()
```

```python
income_data.head(1)
```

```python
income_data['FIPS'] = income_data['Geography'].str[-5:] #getting the last 5 numbers which is the FIPS code
income_data.head(10)
```

```python
income_data = income_data.rename(columns={'Unnamed: 3': 'State Name'})
income_data.head(10)
```

```python
column_names = income_data.columns.tolist()

'Geography' in column_names, column_names[:10]
```

```python
file_path_v = '../data/Vaccination_Coverage_among_Adults__18__Years__20240107.csv'
vaccination_data = pd.read_csv(file_path_v)
vaccination_data.head()
```

```python
income_data['FIPS'] = income_data['FIPS'].astype(str)
vaccination_data['FIPS'] = vaccination_data['FIPS'].astype(str)
vaccination_data['Estimate (%)'] = pd.to_numeric(vaccination_data['Estimate (%)'], errors='coerce') #added this and next line because scatter plot wasn't working due to type error
vaccination_data['Estimate (%)'] = vaccination_data['Estimate (%)'].astype(float)
```

```python
matching_fips_df = pd.merge(income_data, vaccination_data, on = 'FIPS', how='inner') #joining them

#matching_fips = matching_fips_df['FIPS'].unique()
#print(matching_fips)
```

```python
num_rows = matching_fips_df.shape[0]
print(num_rows)
```

```python
plt.figure(figsize=(100, 60))
plt.scatter(year_filtered_fips['Estimate!!Households!!Median income (dollars)'], year_filtered_fips['Estimate (%)'])
plt.title('Scatter Plot of Median Household Income vs. Vaccination Rate')
plt.xlabel('Median Household Income ($)')
plt.ylabel('Vaccination Rate (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(year_filtered_fips['Estimate!!Households!!Median income (dollars)'], year_filtered_fips['Estimate (%)'])

plt.yticks(np.arange(0, 101, 5))  #makes it go by 5s

plt.title('Scatter Plot of Median Household Income vs. Vaccination Rate')
plt.xlabel('Median Household Income ($)')
plt.ylabel('Vaccination Rate (%)')
plt.grid(True)

plt.ylim(0, 100) #makes y axsis from 0 to 100

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(year_filtered_fips['FIPS'], year_filtered_fips['Estimate (%)'])

plt.yticks(np.arange(0, 101, 5))  #makes it go by 5s

plt.title('Scatter Plot of Median Household Income vs. Vaccination Rate')
plt.xlabel('FIPS')
plt.ylabel('Vaccination Rate (%)')
plt.grid(True)

plt.ylim(0, 100) #makes y axsis from 0 to 100

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
plt.plot(year_filtered_fips['FIPS'], year_filtered_fips['Estimate (%)'], marker='o', linestyle='-')

plt.yticks(np.arange(0, 101, 5))  #makes it go by 5s

plt.title('Line Plot of Median Household Income vs. Vaccination Rate')
plt.xlabel('FIPS')
plt.ylabel('Vaccination Rate (%)')
plt.grid(True)

plt.ylim(0, 100) #makes y axsis from 0 to 100

plt.show()
```
